```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sin_position_embedding(pos, i, d_model):
    # Tính toán giá trị của hàm sin
    value = np.sin(pos / (10000 ** (2 * i / d_model)))

    # Vẽ đồ thị
    plt.figure(figsize=(10, 6))
    plt.plot(pos, value, label=f"i={i}, d_model={d_model}")
    plt.xlabel("Position (pos)")
    plt.ylabel("Value")
    plt.title(r"$\sin\left(\frac{pos}{10000^{\frac{2i}{d_{model}}}}\right)$")
    plt.legend()
    plt.grid(True)
    plt.show()

# Tham số
pos = np.arange(0, 50, 0.1)  # Tập hợp các giá trị pos từ 0 đến 50 với bước nhảy 0.1
i = 0  # Chỉ số chiều
d_model = 512  # Kích thước của embedding

# Gọi hàm để vẽ đồ thị
plot_sin_position_embedding(pos, i, d_model)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_position_embedding_sin(pos, i_values, dmodel):
    plt.figure(figsize=(12, 8))
    
    for i in i_values:
        position_values = np.sin(pos / (10000 ** (2 * i / dmodel)))
        plt.plot(pos, position_values, label=f'sin(pos / 10000^(2*{i}/{dmodel}))')
    
    plt.xlabel('Position (pos)')
    plt.ylabel('Value')
    plt.title(f'Position Encoding for i={i_values} and d_model={dmodel}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Tham số
pos = np.arange(0, 50, 1)  # Các vị trí từ 0 đến 49
i_values = [0, 2, 4, 6]    # Các giá trị i cần vẽ
dmodel = 50               # Kích thước của embedding

# Vẽ đồ thị
plot_position_embedding_sin(pos, i_values, dmodel)
```
